```python
import pandas as  pd
```

```python
dataset = pd.read_excel("/content/Data - Survey Monkey Output.xlsx", sheet_name = 'Questions')
dataset
```

```python
dataset_modified = dataset.copy() #making a copy of the dataset
dataset_modified.columns
```

```python
drop_col = ['Start Date', 'End Date', 'Email Address',
            'First Name', 'Last Name', 'Custom Data 1', ]
drop_col #making a list to drop columns easily
```

```python
dataset_modified = dataset_modified.drop(columns = drop_col)
dataset_modified #dropping empty demographic data
```

```python
id_vars = (dataset_modified.columns[:8]) #creating a variable for melt functon
dataset_melted = dataset_modified.melt(id_vars = id_vars, value_vars = None, var_name = 'Questions+responses', value_name = 'Answers') #pivoting questions, responses and answers
dataset_melted
```

```python
questions = pd.read_excel("/content/Data - Survey Monkey Output.xlsx", sheet_name = 'Questions_Transpose')
questions #need to change the unnamed columns to drop them? can use iloc here or indexing based drop like questions[0]
```

```python
questions = questions.drop(columns = ['Unnamed: 0', 'Unnamed: 1'])
questions
```

```python
dataset_merged = pd.merge (left=dataset_melted, right=questions, how='left', left_on= "Questions+responses", right_on= "Questions+responses")
print ('Org_dataset', len(dataset_melted))
print ('New_dataset', len(dataset_merged))
dataset_merged
```

```python
respondents = dataset_merged[dataset_merged["Answers"].notna()] #checks for non missing value
respondents = respondents.groupby("Questions")["Respondent ID"].nunique().reset_index() #reset_index() doesn't convert the dataframe into series
respondents.rename(columns= {"Respondent ID": "Responses"}, inplace=True)
respondents
```

```python
dataset_merged_new = pd.merge (left=dataset_merged, right=respondents, how='left', left_on= "Questions", right_on= "Questions")
print ('Org_dataset', len(dataset_merged))
print ('New_dataset', len(dataset_merged_new))
dataset_merged_new.rename(columns= {"Responses_x": "original_responses", "Responses_y": "filtered_responses"}, inplace=True)
dataset_merged_new
```

```python
same_answers = dataset_merged
same_answers = same_answers.groupby(['Questions+responses', 'Answers'])["Respondent ID"].nunique().reset_index()
same_answers.rename(columns= {"Respondent ID": "Same Answers"}, inplace= True)
same_answers
```

```python
datasetmerged_new = pd.merge (left=dataset_merged_new, right=same_answers, how='left', left_on= ["Questions+responses", "Answers"], right_on= ["Questions+responses", "Answers"])
print ('Org_dataset', len(dataset_merged_new))
print ('New_dataset', len(datasetmerged_new))
datasetmerged_new
```

```python
datasetmerged_new["Same Answers"] = datasetmerged_new["Same Answers"].fillna(0)
datasetmerged_new
```

```python
datasetmerged_new.to_excel("/SurveyMonkey_final.xlsx", index= False)
```
